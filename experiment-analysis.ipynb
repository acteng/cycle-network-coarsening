```python
with open('data/bike_curve_outputs.pkl', 'rb') as f:
    metrics1 = pickle.load(f)
    
with open('data/bike_curve_outputs_2.pkl', 'rb') as f:
    metrics2 = pickle.load(f)
```

```python
columns = ['hypervolume','diversity','signed area','avg_trade_off_rate','num routes','distance','cycling_uptake','walking_uptake']
```

```python
results_list = []

for oa in list(metrics1.keys()):
    print(oa)
    
    if len(metrics1[oa]) > 0:
        if len(metrics1[oa]['change vars']) > 0:
            curve_vars = metrics1[oa]['bike curve'].mean(axis = 0)
            change_vars = metrics1[oa]['change vars']

            results_append = {}
            results_append['hypervolume'] = curve_vars[0]
            results_append['diversity'] = curve_vars[1]
            results_append['signed area'] = curve_vars[2]
            results_append['avg_trade_off_rate'] = curve_vars[3]
            results_append['num routes'] = curve_vars[4]
            results_append['distance'] = curve_vars[5]
            results_append['cycling_uptake'] = change_vars[0][0]
            results_append['walking_uptake'] = change_vars[0][1]

            results_list.append(results_append)
```

```python
results_pd = pd.DataFrame(results_list)
```

```python
results_pd.columns
```

```python
results_pd['num_routes_norm'] = results_pd['num routes'] / results_pd['distance']
```
